```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SEED = 1234
rng = np.random.default_rng(seed=SEED)
```

```python
eta = np.array([0.4, 0.6])
mu = np.array([-1.0, 1.0])
sigma = np.array([0.6, 0.5])
```

```python
SIZE = 10000
k = rng.choice(len(eta), size=SIZE, p=eta)
x = rng.normal(loc=mu[k], scale=sigma[k], size=SIZE)
```

```python
(fig, ax) = plt.subplots()

# 頻度分布をプロット
ax.hist(x, bins="auto", density=True, edgecolor="black")

# 確率密度関数をプロット
x_ = np.linspace(*ax.get_xlim(), 100)
y_ = np.dot(scipy.stats.norm.pdf(x_[:, np.newaxis], loc=mu, scale=sigma), eta)
ax.plot(x_, y_)

plt.show()
```

```python
np.save("gmm_data.npy", x)
```
