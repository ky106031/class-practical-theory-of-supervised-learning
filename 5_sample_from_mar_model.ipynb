```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SEED = 1234
rng = np.random.default_rng(seed=SEED)
```

```python
eta = np.array([0.5, 0.5])

c = np.array([0.0, 0.0])
phi = np.array([[0.5], [1.1]])
sigma = np.array([5.0, 1.0])

def mar_func(x):
    k = rng.choice(len(eta), p=eta)
    return c[k] + np.dot(phi[k], np.flip(x)) + rng.normal(scale=sigma[k])
```

```python
p = len(phi[0])
t0 = 100
T = 200
x = rng.normal(scale=sigma[0], size=p).tolist()
for _ in range(t0 + T):
    x.append(mar_func(x[-p:]))
x = x[t0:]
```

```python
(fig, ax) = plt.subplots()
ax.plot(x)
ax.grid()
plt.show()
```

```python
np.save("mar_data.npy", x)
```
