```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SEED = 1234
rng = np.random.default_rng(seed=SEED)
```

```python
c = 0.0
phi = [1.5, -0.75]
sigma = 1.0

def ar_func(x):
    return c + np.dot(phi, np.flip(x)) + rng.normal(scale=sigma)
```

```python
t0 = 100
T = 144
x = rng.normal(scale=sigma, size=len(phi)).tolist()
for _ in range(t0 + T):
    x.append(ar_func(x[-len(phi):]))
x = x[t0:]
```

```python
(fig, ax) = plt.subplots()
ax.plot(x)
ax.set_xticks(np.arange(0, ax.get_xlim()[1], 12))
ax.grid(axis="x")
plt.show()
```

```python
np.save("ar_model_data.npy", x)
```
